## Function to Generate Patches
Stubs for NDVI, NDBI and creating 16x16 arrays.


```python
def calculate_indices(nir, red, swir1):
    # NDVI = (NIR - RED) / (NIR + RED)
    ndvi = (nir - red) / (nir + red + 1e-8)
    # NDBI = (SWIR1 - NIR) / (SWIR1 + NIR + 1e-8)
    ndbi = (swir1 - nir) / (swir1 + nir + 1e-8)
    return ndvi, ndbi

# Note: To fully run this, place your local USGS TIF files and downloaded ESRI LULC TIF into logic here.
print('Patch extraction logic ready.')
```
