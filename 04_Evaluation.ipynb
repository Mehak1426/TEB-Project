# 4. Evaluation
Calculate Confusion Matrix and Intersection over Union (IoU).


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, jaccard_score
```

## Confusion Matrix and IoU


```python
# y_true = ... 
# y_pred = model.predict(x_test).argmax(axis=1)

def evaluate_model(y_true, y_pred, class_names):
    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
    fig, ax = plt.subplots(figsize=(10,10))
    disp.plot(ax=ax, cmap='Blues')
    plt.title('Confusion Matrix')
    plt.show()
    
    # IoU
    iou = jaccard_score(y_true, y_pred, average=None)
    for i, score in enumerate(iou):
        print(f'IoU for class {class_names[i]}: {score:.4f}')
    print(f'Mean IoU: {np.mean(iou):.4f}')

print('Evaluation metrics defined.')
```
